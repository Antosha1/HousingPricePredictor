```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import statsmodels.api as sm

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/clean_data.csv')
df.head()
```

```python
y = df['SalePrice'].apply(lambda x: np.log(x))
X = df.drop(columns=['SalePrice'])
```

Заранее выделим числовые признаки.

```python
num_features = []

for feature in list(X):
    if X[feature].nunique() > 10:
        num_features.append(feature)
        
print(*num_features, sep=', ')
```

```python
ss = StandardScaler()
X[num_features] = ss.fit_transform(X[num_features])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
X_train[num_features]
```

```python
X_1 = sm.add_constant(X_train)
model = sm.OLS(y_train, X_1)
results = model.fit()
```

```python
print(results.summary())
```

```python
X_test = sm.add_constant(X_test)
preds = results.predict(X_test)
```

```python
round(sum((preds - y_test)**2) / len(preds), 3)
```

```python
preds = preds.reset_index().iloc[:, 1:].apply(lambda x: np.exp(x))
y_test = y_test.reset_index().iloc[:, 1:].apply(lambda x: np.exp(x))

plt.figure(figsize=(15, 12))

plt.plot(y_test, color='b')
```
